```python
import pandas as pd
import os
```

```python
path = 'lab/'
list_of_files = os.listdir(path)
```

```python
df = pd.read_csv(path+list_of_files[0], sep='\t')
for table_name in list_of_files[1:]:
    df = df.append(pd.read_csv(path+table_name, sep='\t'))
```

```python
cols = ['Регистрационный_номер',  'Номер_эпизода', 'Код_теста',\
       'Синоним', 'Наименование', 'Ед_изм', 'Значение', 'Значение_число', \
        'Метод_теста','Анализатор_рабочий_лист', 'Нормы_для_теста', 'Флаг_результата_теста']
df = df[cols]
```

```python
isn = df.isnull().sum()
isn[isn>0]
```

```python
drop_cols = list(isn[isn>15000].index)
df.drop(drop_cols, axis=1, inplace=True)
df.dropna(inplace=True)
```
